```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import random
import pandas as pd



# 사용자명과 비밀번호를 파일에서 읽어오는 함수
def read_credentials():
    with open('data/id.txt', 'r', encoding='utf-8') as f:
        username = f.read().strip()  # 파일에서 읽은 사용자명을 반환 (줄바꿈 제거)
    
    with open('data/password.txt', 'r', encoding='utf-8') as f:
        password = f.read().strip()  # 파일에서 읽은 비밀번호를 반환 (줄바꿈 제거)
    
    return username, password


driver = webdriver.Chrome()
url= 'https://www.instagram.com/'
driver.get(url)

# 사용자명과 비밀번호 읽어오기
username, password = read_credentials()

# 사용자 이름과 비밀번호 입력 필드 찾기
login_id = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, '#loginForm > div.x9f619.xjbqb8w.x78zum5.x168nmei.x13lgxp2.x5pf9jr.xo71vjh.xqui205.x1n2onr6.x1plvlek.xryxfnj.x1c4vz4f.x2lah0s.xdt5ytf.xqjyukv.x1qjc9v5.x1oa3qoh.x1nhvcw1 > div:nth-child(1) > div > label > input'))
    )
login_id.click()
login_id.send_keys(username)

login_password = driver.find_element(By.CSS_SELECTOR,'#loginForm > div.x9f619.xjbqb8w.x78zum5.x168nmei.x13lgxp2.x5pf9jr.xo71vjh.xqui205.x1n2onr6.x1plvlek.xryxfnj.x1c4vz4f.x2lah0s.xdt5ytf.xqjyukv.x1qjc9v5.x1oa3qoh.x1nhvcw1 > div:nth-child(2) > div > label > input')  # 비밀번호 입력 필드
login_password.click()
login_password.send_keys(password, Keys.ENTER)

time.sleep(10)

# 로그인 성공 후 특정 URL로 이동
driver.get("https://www.instagram.com/hanwhaeagles_soori/")
driver.maximize_window()

def scroll_down(driver, max_scrolls=56):
    last_height = driver.execute_script("return document.body.scrollHeight")
    scroll_count = 0  # 스크롤 횟수를 추적하기 위한 변수
    
    while scroll_count < max_scrolls:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight)')
        time.sleep(random.uniform(7,10))
        current_height = driver.execute_script("return document.body.scrollHeight")
        
        if last_height == current_height and scroll_count == max_scrolls - 1:
            break  # 스크롤이 더 이상 진행되지 않으면 종료
        
        last_height = current_height
        scroll_count += 1  # 스크롤 횟수 증가
        

def scroll_up_and_find_element(driver, max_scrolls=10):
    scroll_count = 0
    
    while scroll_count < max_scrolls:
        try:
            # 페이지 스크롤을 올리기
            driver.execute_script("window.scrollBy(0, -window.innerHeight);")
            time.sleep(2)  # 페이지가 로드되는 시간을 대기
            
            # 요소가 화면에 보이면 클릭
            click_page = driver.find_element(By.CSS_SELECTOR, '._ac7v.x1f01sob.xcghwft.xat24cr.xzboxd6 > .x1lliihq.x1n2onr6.xh8yej3.x4gyw5p.x1ntc13c.x9i3mqj.x11i5rnm.x2pgyrj')  
            click_page.click()
            print("요소를 클릭했습니다.")
            break  # 요소를 클릭한 후 루프 종료
        
        except:
            # 요소를 찾을 수 없으면 계속 스크롤
            print(f"스크롤 {scroll_count + 1}에서 요소를 찾을 수 없습니다.")
            scroll_count += 1  # 스크롤 횟수 증가



def click_more_button():
    # 스크롤을 끝까지 내리며 "더보기" 버튼을 찾기
    max_scroll_attempts = 3  # 더보기 버튼을 못 찾으면 최대 3번 스크롤
    scroll_attempts = 0
    more_button_found = False

    while True:
        # 페이지 끝까지 스크롤
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.uniform(1,3))  # 스크롤 후 잠시 대기

        try:
            # "더보기" 버튼 찾기
            more_button = driver.find_element(By.CSS_SELECTOR, 'body > div.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div.xb88tzc.xw2csxc.x1odjw0f.x5fp0pe.x1qjc9v5.xjbqb8w.x1lcm9me.x1yr5g0i.xrt01vj.x10y3i5r.xr1yuqi.xkrivgy.x4ii5y1.x1gryazu.x15h9jz8.x47corl.xh8yej3.xir0mxb.x1juhsu6 > div > article > div > div.x1qjc9v5.x972fbf.xcfux6l.x1qhh985.xm0m39n.x9f619.x78zum5.xdt5ytf.x1iyjqo2.x5wqa0o.xln7xf2.xk390pu.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.x65f84u.x1vq45kp.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x1n2onr6.x11njtxf > div > div > div.x78zum5.xdt5ytf.x1q2y9iw.x1n2onr6.xh8yej3.x9f619.x1iyjqo2.x18l3tf1.x26u7qi.xy80clv.xexx8yu.x4uap5.x18d9i69.xkhd6sd > div.x78zum5.xdt5ytf.x1iyjqo2.xs83m0k.x2lwn1j.x1odjw0f.x1n2onr6.x9ek82g.x6ikm8r.xdj266r.x11i5rnm.x4ii5y1.x1mh8g0r.xexx8yu.x1pi30zi.x18d9i69.x1swvt13 > ul > div:nth-child(3) > div > div > li > div > button > div > svg')  # 더보기 버튼
            more_button.click()  # 더보기 버튼 클릭
            time.sleep(random.uniform(1,3))  # 버튼 클릭 후 대기
            print("더보기 버튼 클릭됨")
            more_button_found = True
            scroll_attempts = 0
        except:
            print("더보기 버튼 없음, 계속 스크롤 중...")
            scroll_attempts += 1  # 스크롤 시도 횟수 증가

        # "더보기" 버튼을 못 찾았고, 시도 횟수가 최대에 도달했을 때 한 번 더 스크롤 후 댓글 추출 시작
        if scroll_attempts == max_scroll_attempts - 1:
            print("마지막 한 번 더 페이지를 끝까지 스크롤합니다.")
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(random.uniform(1, 3))  # 페이지 끝까지 스크롤 후 잠시 대기

        # "더보기" 버튼이 더 이상 없으면 댓글을 추출하고 반복문 종료
        if scroll_attempts >= max_scroll_attempts:
            print("더보기 버튼이 더 이상 없습니다. 댓글을 추출합니다.")
            break  # 더이상 "더보기" 버튼이 없으면 종료




element_data = []

def extract_comments():

   
    while True:
        
        try:
            # 본문 텍스트 추출
            body_content = driver.find_element(By.CSS_SELECTOR, 'body > div.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div.xb88tzc.xw2csxc.x1odjw0f.x5fp0pe.x1qjc9v5.xjbqb8w.x1lcm9me.x1yr5g0i.xrt01vj.x10y3i5r.xr1yuqi.xkrivgy.x4ii5y1.x1gryazu.x15h9jz8.x47corl.xh8yej3.xir0mxb.x1juhsu6 > div > article > div > div.x1qjc9v5.x972fbf.xcfux6l.x1qhh985.xm0m39n.x9f619.x78zum5.xdt5ytf.x1iyjqo2.x5wqa0o.xln7xf2.xk390pu.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.x65f84u.x1vq45kp.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x1n2onr6.x11njtxf > div > div > div.x78zum5.xdt5ytf.x1q2y9iw.x1n2onr6.xh8yej3.x9f619.x1iyjqo2.x18l3tf1.x26u7qi.xy80clv.xexx8yu.x4uap5.x18d9i69.xkhd6sd > div.x78zum5.xdt5ytf.x1iyjqo2.xs83m0k.x2lwn1j.x1odjw0f.x1n2onr6.x9ek82g.x6ikm8r.xdj266r.x11i5rnm.x4ii5y1.x1mh8g0r.xexx8yu.x1pi30zi.x18d9i69.x1swvt13 > ul > div.x1qjc9v5.x972fbf.xcfux6l.x1qhh985.xm0m39n.x9f619.x78zum5.xdt5ytf.x2lah0s.xk390pu.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x1n2onr6.xggy1nq.x11njtxf > li > div > div > div._a9zr > div._a9zs > h1').text
            

            # '매진' 또는 '이벤트'가 본문 텍스트에 포함되어 있는지 확인
            if '매진' in body_content or '이벤트' in body_content or '사인회' in body_content:
                print("본문에 '매진' 또는 '이벤트' 또는 '사인회'가 포함되어 있습니다. 댓글 추출 시작.")
                
                click_more_button()

                # 댓글 추출 시작
                for n in range(1, 3001):
                    try:
                        # n에 해당하는 댓글 요소 선택
                        element = driver.find_element(By.CSS_SELECTOR, f'body > div.x14dbnvc.x67yw2k.x3hiddl.x1xb1xrg.xz3gdfk.xbi9o00.x1xkblxv.x4666fc.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div.xb88tzc.xw2csxc.x1odjw0f.x5fp0pe.x1qjc9v5.xjbqb8w.x1lcm9me.x1yr5g0i.xrt01vj.x10y3i5r.xr1yuqi.xkrivgy.x4ii5y1.x1gryazu.x15h9jz8.x47corl.xh8yej3.xir0mxb.x1juhsu6 > div > article > div > div.x1qjc9v5.x972fbf.xcfux6l.x1qhh985.xm0m39n.x9f619.x78zum5.xdt5ytf.x1iyjqo2.x5wqa0o.xln7xf2.xk390pu.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.x65f84u.x1vq45kp.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x1n2onr6.x11njtxf > div > div > div.x78zum5.xdt5ytf.x1q2y9iw.x1n2onr6.xh8yej3.x9f619.x1iyjqo2.x18l3tf1.x26u7qi.xy80clv.xexx8yu.x4uap5.x18d9i69.xkhd6sd > div.x78zum5.xdt5ytf.x1iyjqo2.xs83m0k.x2lwn1j.x1odjw0f.x1n2onr6.x9ek82g.x6ikm8r.xdj266r.x11i5rnm.x4ii5y1.x1mh8g0r.xexx8yu.x1pi30zi.x18d9i69.x1swvt13 > ul > div:nth-child(3) > div > div > div:nth-child({n}) > ul > div > li > div > div > div._a9zr > div._a9zs > span')
                        date = driver.find_element(By.CSS_SELECTOR, 'body > div.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div.xb88tzc.xw2csxc.x1odjw0f.x5fp0pe.x1qjc9v5.xjbqb8w.x1lcm9me.x1yr5g0i.xrt01vj.x10y3i5r.xr1yuqi.xkrivgy.x4ii5y1.x1gryazu.x15h9jz8.x47corl.xh8yej3.xir0mxb.x1juhsu6 > div > article > div > div.x1qjc9v5.x972fbf.xcfux6l.x1qhh985.xm0m39n.x9f619.x78zum5.xdt5ytf.x1iyjqo2.x5wqa0o.xln7xf2.xk390pu.xdj266r.x11i5rnm.xat24cr.x1mh8g0r.x65f84u.x1vq45kp.xexx8yu.x4uap5.x18d9i69.xkhd6sd.x1n2onr6.x11njtxf > div > div > div.x78zum5.xdt5ytf.x1q2y9iw.x1n2onr6.xh8yej3.x9f619.x1iyjqo2.x18l3tf1.x26u7qi.xy80clv.xexx8yu.x4uap5.x18d9i69.xkhd6sd > div.x1yztbdb.x1h3rv7z.x1swvt13 > div > div > a > span > time').text
                        # 댓글 데이터 리스트에 추가
                        element_data.append({'날짜': date,'댓글 번호': n, '댓글': element.text})
                        print(f'Found comment {n}')
                    
                    except Exception as e:
                        print(f'Comment {n} not found, stopping.')
                        continue

                # "다음 글" 버튼 클릭
                try:
                    time.sleep(random.uniform(2, 4))
                    next_button = driver.find_element(By.CSS_SELECTOR, 'body > div.x14dbnvc.x67yw2k.x3hiddl.x1xb1xrg.xz3gdfk.xbi9o00.x1xkblxv.x4666fc.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div:nth-child(1) > div > div > div._aaqf._aaqh > button > div > span > svg')  # "다음 글" 버튼
                    next_button.click()  # 버튼 클릭
                    print("다음 글 버튼 클릭됨")
                    time.sleep(random.uniform(2, 4))  # 다음 글 로딩 대기  
                except Exception as e:
                    print(f"다음 글 버튼을 찾을 수 없습니다: {e}")
                    break  # 여기에서 루프 종료

            else:
                print("본문에 '매진' 또는 '이벤트' 또는 '사인회'가 포함되지 않았습니다. 다음 글로 이동.")

                # "다음 글" 버튼 찾기 및 클릭
                try:
                    time.sleep(random.uniform(2, 4))
                    next_button = driver.find_element(By.CSS_SELECTOR, 'body > div.x14dbnvc.x67yw2k.x3hiddl.x1xb1xrg.xz3gdfk.xbi9o00.x1xkblxv.x4666fc.x1n2onr6.xzkaem6 > div.x9f619.x1n2onr6.x1ja2u2z > div > div.x1uvtmcs.x4k7w5x.x1h91t0o.x1beo9mf.xaigb6o.x12ejxvf.x3igimt.xarpa2k.xedcshv.x1lytzrv.x1t2pt76.x7ja8zs.x1n2onr6.x1qrby5j.x1jfb8zj > div > div > div > div > div:nth-child(1) > div > div > div._aaqf._aaqh > button > div > span > svg')  # "다음 글" 버튼
                    next_button.click()  # 버튼 클릭
                    print("다음 글 버튼 클릭됨")
                    time.sleep(random.uniform(2, 4))  # 다음 글 로딩 대기  
                                  
                except Exception as e:
                    print(f"다음 글 버튼을 찾을 수 없습니다: {e}")
                    break 

        except Exception as e:
            print(f"기타 오류 발생: {e}")
            break





scroll_down(driver)
scroll_up_and_find_element(driver)
extract_comments()

```
