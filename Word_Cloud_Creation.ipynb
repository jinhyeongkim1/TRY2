```python
import re

def clean_text(text):
    # '@아이디' 형식을 찾고 이를 제거
    text = re.sub(r'@\S+', '', text)  # '@'로 시작하고 알파벳과 숫자로 이루어진 아이디 제거
    # 이모티콘 및 특수 문자 제거 (한글, 영어, 숫자, 공백만 남기기)
    text = re.sub(r'[^\w\s,가-힣]', '', text)  # 한글, 영어, 숫자, 공백, 쉼표만 남기기
    text = re.sub(r'[ㅋㅎㅣㅉㅠㅜ]', '', text)
    # 줄바꿈 (\n)과 탭 (\t)을 공백으로 바꾸기
    text = re.sub(r'[\n\t]', ' ', text)
    text = text.strip()  # 문자열 양옆의 공백을 제거
    return text
```

```python
# 특정 단어들 합치기
def merge_tokens(tokens):
    merged_tokens = []
    i = 0
    while i < len(tokens):
        # '노시'와 '환' 합치기
        if i < len(tokens) - 1 and tokens[i] == '노시' and tokens[i+1] == '환':
            merged_tokens.append('노시환')
            i += 2  # '노시환'은 하나로 합쳤으므로 2칸 건너뛰기
        # '썸'과 '폼' 합치기
        elif i < len(tokens) - 1 and tokens[i] == '썸' and tokens[i+1] == '폼':
            merged_tokens.append('썸니폼')
            i += 2
        # '환'이 단독으로 있을 때 '노시환'으로 변환
        elif tokens[i] == '환':
            merged_tokens.append('노시환')
            i += 1  # '환'은 하나만 건너뛰기
        # 그 외의 경우 그냥 추가
        else:
            merged_tokens.append(tokens[i])
            i += 1
    return merged_tokens
```

```python
# '최원호'로 바꾸기 위한 단어들
target_words = ['최', '원호', '워노', '최씨', '최원']

# '최원호'로 바꾸는 함수
def merge_tokens(tokens):
    merged_tokens = []
    for token in tokens:
        if token in target_words:
            merged_tokens.append('최원호')
        else:
            merged_tokens.append(token)
    return merged_tokens
```

```python
def merge_tokens(tokens):
    merged_tokens = []
    i = 0
    while i < len(tokens):
        # '장진'과 '혁'이 함께 있을 경우 '장진혁'으로 바꾸기
        if i < len(tokens) - 1 and tokens[i] == '장진' and tokens[i+1] == '혁':
            merged_tokens.append('장진혁')
            i += 2  # 두 단어를 하나로 합치므로 두 칸 건너뜀
            continue
        # '황영'과 '묵'이 함께 있을 경우 '황영묵'으로 바꾸기
        elif i < len(tokens) - 1 and tokens[i] == '황영' and tokens[i+1] == '묵':
            merged_tokens.append('황영묵')
            i += 2  # 두 단어를 하나로 합치므로 두 칸 건너뜀
            continue
        # '묵'과 '묵이베츠'가 함께 있을 경우 '황영묵'으로 바꾸기
        elif tokens[i] == '묵' or tokens[i] == '묵이베츠':
            merged_tokens.append('황영묵')
            i += 1  # 하나의 '묵'을 '황영묵'으로 바꿨으므로 한 칸만 건너뜀
            continue
        # 그 외의 경우는 그대로 추가
        merged_tokens.append(tokens[i])
        i += 1
    return merged_tokens

# 각 댓글에서 명사를 추출한 후, 규칙에 맞게 합치기
merged_tokens_list = [merge_tokens(tokens) for tokens in tokens_list]
```
